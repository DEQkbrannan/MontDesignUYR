Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('UpperYaquinaBacteria.csv')
df.head()
```

```python
df.dtypes
```

```python
df['STATION'] = df['STATION'].astype(str)
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df.dtypes
```

```python
median_value = df.groupby('STATION')['EC_val'].median()
median_value
```

```python
aggregated_df = df.groupby('STATION')['EC_val'].agg(['median', 'mean', 'max', 'min'])
aggregated_df
```

```python
df.boxplot(column='EC_val', by='STATION')
```

```python
# Assuming df is your DataFrame
df.boxplot(column='EC_val', by='STATION')

# Set the y-axis to logarithmic scale
plt.gca().set_yscale('log')

# Customizations
plt.title('Boxplot of E. coli levels by Station (Log Scale)')
plt.suptitle('')  # Suppresses the default title to avoid duplication
plt.xlabel('Station')
plt.ylabel('E .coli (cpu/100 ml) (log scale)')

# Add a horizontal line at y = 406
plt.axhline(y=406, color='r', linestyle='-')

# Show the plot
plt.show()
```
